```python
import xad.adj_1st as xadj
```

```python
x0_ad = xadj.Real(1.0)
x1_ad = xadj.Real(1.5)
x2_ad = xadj.Real(1.3)
x3_ad = xadj.Real(1.2)
```

```python
%%time

with xadj.Tape() as tape:
    tape.registerInput(x0_ad)
    tape.registerInput(x1_ad)
    tape.registerInput(x2_ad)
    tape.registerInput(x3_ad)

    tape.newRecording()
    y = 2 * x0_ad + x1_ad - x2_ad * x3_ad

    tape.registerOutput(y)
    y.setAdjoint(1.0)
    tape.computeAdjoints()

    print(f"y = {y.getValue()}")
    print(f"dy/dx0 = {(x0_ad.getDerivative())}")
    print(f"dy/dx1 = {(x1_ad.getDerivative())}")
    print(f"dy/dx2 = {(x2_ad.getDerivative())}")
    print(f"dy/dx3 = {(x3_ad.getDerivative())}")
```
